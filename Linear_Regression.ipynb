```python
import numpy as np
import matplotlib.pyplot as plt

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")

m = len(x_train)

print("number of inputs =", m)

i=0

x_i = x_train[i]
y_i = y_train[i]

print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")

def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples
      w,b (scalar)    : model parameters
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb

tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()

w = 200
b = 100
x_i = 1.2
cost_650sqft = w * x_i + b

print(f"${cost_650sqft:.0f} thousand dollars")




```

Let's re-plot the data with the optimal `w=200` and `b=100` values to see the perfect fit.

```python
# Calculate the model output with the optimal parameters
optimal_f_wb = compute_model_output(x_train, 200, 100)

# Plot our optimal model prediction
plt.figure()
plt.plot(x_train, optimal_f_wb, c='g', label='Optimal Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

# Set the title
plt.title("Housing Prices (Optimal Fit)")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
def compute_cost(x, y, w, b):
    """
    Computes the cost function for linear regression.
    Args:
      x (ndarray (m,)): Data, m examples
      y (ndarray (m,)): Target values
      w,b (scalar)    : model parameters
    Returns
      total_cost (float): The cost of using w,b as the parameters for linear regression
    """
    m = x.shape[0]
    cost_sum = 0
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum = cost_sum + cost
    total_cost = (1 / (2 * m)) * cost_sum
    return total_cost
```

Now, let's calculate the cost for the `w=100, b=100` and `w=200, b=100` values you used previously to see which set of parameters provides a better fit to your training data (`x_train`, `y_train`). A lower cost indicates a better model.

```python
# Cost for w=100, b=100
cost_100_100 = compute_cost(x_train, y_train, 100, 100)
print(f"Cost for w=100, b=100: {cost_100_100}")

# Cost for w=200, b=100
cost_200_100 = compute_cost(x_train, y_train, 200, 100)
print(f"Cost for w=200, b=100: {cost_200_100}")

# Let's try some 'better' values, by visual inspection from the plot, w=200 and b=100 seems closer.
# But for exact fit, for x=1, y=300 and for x=2, y=500
# 300 = w*1 + b
# 500 = w*2 + b
# Subtracting first from second: 200 = w
# Then 300 = 200*1 + b => b = 100
# So, w=200, b=100 should give perfect fit (cost=0) for these two points.
cost_perfect_fit = compute_cost(x_train, y_train, 200, 100)
print(f"Cost for w=200, b=100 (should be optimal for training data): {cost_perfect_fit}")
```
